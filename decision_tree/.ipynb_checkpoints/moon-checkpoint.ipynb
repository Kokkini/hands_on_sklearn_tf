```python
# import
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.datasets import make_moons
from sklearn.model_selection import GridSearchCV
```

```python
X,y = make_moons(n_samples=20000, noise=0.2)
print(X.shape)
print(y.shape)
```

```python
plt.scatter(X[:,0],X[:,1],c=y,s=0.1)
```

```python
train_size = 10000
X_train, X_test, y_train, y_test = X[:train_size], X[train_size:], y[:train_size], y[train_size:]
```

```python
# create classifier
tree_clf = DecisionTreeClassifier(max_depth=5)
tree_clf.fit(X_train,y_train)
```

```python
tree_clf.score(X_test, y_test)
```

```python
export_graphviz(tree_clf,class_names=("0","1"))
```

```python
y_test_pred = tree_clf.predict(X_test)
```

```python
a = np.random.uniform(size=[100000,2])*4-2
```

```python
y_a_pred = tree_clf.predict(a)
```

```python
plt.scatter(a[:,0],a[:,1],c=y_a_pred,s=0.1,)
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_test,s=0.1,)
```

```python
# grid search
param_grid = [{"max_depth":[2,4,8]}]
search = GridSearchCV(tree_clf,param_grid,scoring="accuracy",cv=4)
search.fit(X_train,y_train)
```

```python
search.best_params_
```

```python
best_tree = DecisionTreeClassifier(max_depth=8)
best_tree.fit(X_train, y_train)
best_tree.score(X_test, y_test)
```

```python
y_a_pred = best_tree.predict(a)
plt.scatter(a[:,0],a[:,1],c=y_a_pred,s=0.1)
```
