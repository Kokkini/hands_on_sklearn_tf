```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_mldata,make_swiss_roll,make_checkerboard,make_friedman1,make_blobs
from mpl_toolkits import mplot3d
from sklearn.decomposition import KernelPCA,PCA
from sklearn.manifold import LocallyLinearEmbedding
```

```python
X_roll, y_roll = make_swiss_roll(1000)
mnist = fetch_mldata("MNIST original")
board = make_checkerboard((10,10),2)
X_fried, y_fried = make_friedman1(10000)
X_blob, y_blob = make_blobs(1000,3,centers=2)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection="3d")
ax.scatter3D(X_blob[:,0],X_blob[:,1],X_blob[:,2],c=y_blob)
plt.show()
```

```python
pca = PCA(n_components=0.9)
X_blob_reduced = pca.fit_transform(X_blob)
```

```python
print(X_blob_reduced.shape)
print(X_blob.shape)
pca.explained_variance_ratio_
```

```python
plt.scatter(X_blob_reduced[:,0],X_blob_reduced[:,1],c=y_blob)
```

```python
lle = LocallyLinearEmbedding(n_components=2, n_neighbors=10)
X_roll_reduced = lle.fit_transform(X_roll)
plt.scatter(X_roll_reduced[:,0], X_roll_reduced[:,1],c=y_roll)
```
